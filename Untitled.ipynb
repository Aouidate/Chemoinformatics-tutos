```python
def calculate_applicability_domain(training_data, test_data, k, threshold):
    # Extract descriptors and activity from training data
    train_desc = np.array(training_data.iloc[:, 0:-2])
    training_activity = training_data.iloc[:, -1]
    train_indices = training_data.index.tolist()  # Get the indices of the training data

    # Extract descriptors from test data
    test_desc = np.array(test_data.iloc[:, 0:-2])
    test_indices = test_data.index.tolist()  # Get the indices of the test data

    # Calculate Euclidean distance between each test set molecule and its k-nearest neighbors in the training set
    dist = np.empty((test_desc.shape[0],))
    out_of_domain_indices = {}
    for i in range(dist.shape[0]):
        diff = train_desc - test_desc[i, :]
        euclidean_dist = np.linalg.norm(diff, ord=2, axis=1)
        k_indices = np.argpartition(euclidean_dist, k)[:k]
        dist[i] = np.average(euclidean_dist[k_indices])

        # Check if test molecule is out of applicability domain and add its original index to out_of_domain_indices
        if dist[i] > threshold:
            original_test_index = test_indices[i]
            out_of_domain_indices[original_test_index] = dist[i]

    # Return dictionary of indices from the original test set that are out of applicability domain along with their distances
    return out_of_domain_indices
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

def plot_applicability_domain(training_data, test_data, threshold):
    # Calculate the applicability domain
    out_of_domain_indices = calculate_applicability_domain(training_data, test_data, 10, threshold)

    # Extract descriptors from training and test data
    train_desc = np.array(training_data.iloc[:, 0:-2])
    test_desc = np.array(test_data.iloc[:, 0:-2])

    # Plot the applicability domain
    fig, ax = plt.subplots()

    # Set color or size encoding for additional descriptors (if applicable)
    if train_desc.shape[1] > 2:
        c = train_desc[:, 2]  # Example: Third descriptor used for color encoding
        s = train_desc[:, 3]  # Example: Fourth descriptor used for size encoding
    else:
        c = 'blue'
        s = 20

    for i in range(test_desc.shape[0]):
        if i in out_of_domain_indices:
            ax.scatter(test_desc[i, 0], test_desc[i, 1], c='red', s=50)  # Adjust size and color as needed
            ax.annotate(str(i), (test_desc[i, 0], test_desc[i, 1]), textcoords="offset points", xytext=(0, 10), ha='center')
        else:
            ax.scatter(test_desc[i, 0], test_desc[i, 1], c='blue', s=20)  # Adjust size and color as needed

    # Plot the training set with color or size encoding (if applicable)
    if train_desc.shape[1] > 2:
        ax.scatter(train_desc[:, 0], train_desc[:, 1], c=c, s=s, cmap='viridis', marker='s')  # Adjust size and color as needed
    else:
        ax.scatter(train_desc[:, 0], train_desc[:, 1], color='green', marker='s')

    # Create an ellipse representing the applicability domain
    training_mean = train_desc.mean(axis=0)
    training_std = train_desc.std(axis=0)

    # Adjust the dimensions of the ellipse based on the number of descriptors
    ellipse_dimensions = tuple(threshold * 2 * training_std[i] for i in range(test_desc.shape[1]))

    # Adjust the ellipse color and transparency as needed
    ellipse = Ellipse(xy=training_mean, width=ellipse_dimensions[0], height=ellipse_dimensions[1], fill=False, color='gray', alpha=0.5)

    ax.add_patch(ellipse)

    # Set plot labels
    ax.set_xlabel('Descriptor 1')
    ax.set_ylabel('Descriptor 2')

    # Set legend
    ax.legend(['Applicability Domain', 'Outside AD', 'Within AD', 'Training Set'])

    # Show the plot
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.patches import Ellipse
from sklearn.preprocessing import StandardScaler

def plot_pca_applicability_domain(training_data, test_data, threshold):
    # Extract descriptors from training and test data
    train_desc = np.array(training_data.iloc[:, 0:-1])
    test_desc = np.array(test_data.iloc[:, 0:-1])

    # Standardize the data
    scaler = StandardScaler()
    train_desc_scaled = scaler.fit_transform(train_desc)
    test_desc_scaled = scaler.transform(test_desc)

    # Perform PCA
    pca = PCA(n_components=2)
    train_pca = pca.fit_transform(train_desc_scaled)
    test_pca = pca.transform(test_desc_scaled)

    # Plot the PCA space
    fig, ax = plt.subplots()

    # Plot training set
    ax.scatter(train_pca[:, 0], train_pca[:, 1], color='green', marker='s', label='Training Set')

    # Plot test set
    ax.scatter(test_pca[:, 0], test_pca[:, 1], color='blue', label='Test Set')

    # Calculate threshold values for each principal component
    threshold_values = threshold * np.sqrt(pca.explained_variance_)

    # Plot applicability domain boundaries
    for i in range(len(threshold_values)):
        ellipse = Ellipse(xy=(0, 0), width=2 * threshold_values[i], height=2 * threshold_values[i],
                              angle=np.rad2deg(np.arccos(pca.components_[0, i])), fill=False, color='red')
        ax.add_artist(ellipse)

    # Set plot labels
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')

    # Set legend
    ax.legend()

    # Show the plot
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.patches import Ellipse
from sklearn.preprocessing import StandardScaler

def plot_pca_applicability_domain(training_data, test_data, threshold):
    # Extract descriptors and idexes from training and test data
    train_desc = np.array(training_data.iloc[:, 0:-1])
    train_idx = training_data.index.to_list()
    actual_values_train = np.array(training_data.iloc[:, -1])
    test_desc = np.array(test_data.iloc[:, 0:-1])
    test_idx = test.index.to_list()

    # Standardize the data
    scaler = StandardScaler()
    train_desc_scaled = scaler.fit_transform(train_desc)
    test_desc_scaled = scaler.transform(test_desc)

    # Perform PCA
    pca = PCA(n_components=2)
    train_pca = pca.fit_transform(train_desc_scaled)
    test_pca = pca.transform(test_desc_scaled)

    # Plot the PCA space
    fig, ax = plt.subplots(figsize=(10, 8))

    # Plot training set
    im = ax.scatter(train_pca[:, 0], train_pca[:, 1], marker='D', label='Training Set', c = actual_values_train, cmap=plt.cm.jet)

    # Plot test set
    im = ax.scatter(test_pca[:, 0], test_pca[:, 1], color='r', label='Test Set',marker= r'$\clubsuit$', s=150)

    # Calculate threshold values for each principal component
    threshold_values = threshold * np.sqrt(pca.explained_variance_)
    num_pcs = min(len(threshold_values), 2)  # Limit to maximum 2 PCs

    # Plot applicability domain boundaries
    for i in range(num_pcs):
        ellipse = Ellipse(xy=(0, 0), width=2 * threshold_values[i], height=2 * threshold_values[i],
                          angle=np.rad2deg(np.arccos(pca.components_[0, i])), fill=False, color='red')
        ax.add_artist(ellipse)

    # Identify the indices of the outliers in the test set
    outlier_indices = {}
    for i, desc in zip(test_idx, test_desc_scaled):
        is_outlier = True
        for j, train_desc in zip(train_idx, train_desc_scaled):
            dist = np.linalg.norm(desc - train_desc)
            if (dist <= threshold_values).any():
                is_outlier = False
                break
        if is_outlier:
            outlier_indices[i] = True
    # Add molecule indexes for compounds outside the applicability domain
    for index, _ in outlier_indices.items():
        ax.annotate(str(index), (test_pca[index, 0], test_pca[index, 1]), textcoords="offset points",
                    xytext=(0, 10), ha='center', color='red')

    # Set plot labels
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    fig.colorbar(im, ax=ax)
    # Set legend
    ax.legend()
    
    #Set title 
    plt.title('PCA Applicability Domain')

    # Show the plot
    plt.show()
```
